# Ácido-base  monoprótico

## Ácido-base forte

```python
def tit_a(vtit):
    '''Curva de titulação ácida'''
    Ca=0.05 #definindo as constantes
    Cb=0.1
    V0=50
    Kw=1e-14
    b=-(Ca*V0-Cb*vtit)/(vtit+V0)
    H=(-b+raiz(b**2+4*Kw))/2
    return round(-log(H),2)

tit_a(25)
```

```python
def curva():
    Ca=0.05 #definindo as constantes
    Cb=0.1
    V0=20
    Kw=1e-14
    pH,Vtit=[],[] #criando as listas com valores de Vtit e pH
    for vtit in arange(0,20,0.01): #calculando o pH para cada Vtit
        Vtit.append(vtit)
        b=-(Ca*V0-Cb*vtit)/(vtit+V0)
        H=(-b+raiz(b**2+4*Kw))/2
        pH.append(-log(H))
    return pH,Vtit

#plotando e salvando
pH,Vtit=curva()
plt.plot(Vtit,pH,linewidth=1,color='black')
plt.annotate('$[H^+]=\\frac{-b+\sqrt{b^2+4K_w}}{2}$',xy=(0,10),size=20)
plt.annotate('$b=-\\frac{C_a V_{0}-C_b V_{tit}}{V_{tit}+V_0}$',xy=(0,7.5),size=20)
plt.xlabel('$V_{tit}$',size=15)
plt.ylabel('pH',size=15)
plt.title('Curva de titulação de um ácido forte por uma base forte',size=18)
#plt.savefig('Curva de Titulação',dpi=300,format='jpeg')
```

## Ácido-base fraco

```python
#definindo o método numérico para achar raiz 

def sinal(x):
    '''Função sinal. Retorna -1 se x<0, 0 se x=0 e 1 se x>0.'''
    if x>0:
        return 1.0
    elif x<0:
        return -1.0
    else:
        return 0.0

    
def raizbs(f,a,b):
    '''Método da bisseção para achar a raíz da função f no intervalo [a, b].
    Faz a bisseção até que |x2-x1|<=epsilon.'''
    epsilon=(b-a)/10000
    x1=a
    f1=f(x1)
    if f1==0.0:
        return x1
    x2=b
    f2=f(x2)
    if f2==0.0:
        return x2
    while abs(x2-x1)>=epsilon:
        x3=(x1+x2)/2
        f3=f(x3)
        if sinal(f2)!=sinal(f3):
            x1=x3
            f1=f3
        else:
            x2=x3
            f2=f3
    media=(x1+x2)/2
    return media
```

```python
def pH(Va):
    Ca=0.01 #definindo as constantes
    Cb=0.0150
    Vb=25
    Kw=1e-14
    Ka=2.3e-11
    def pHH(ph):
        H=10**(-ph)
        OH=Kw/H
        alpha_bh=H/(H+Ka)
        return H+alpha_bh*((Cb*Vb)/(Va+Vb))-((Ca*Va)/(Va+Vb))-OH
        
    return raizbs(pHH,-log(Ca),14)
```

# Ácido-base diprótico

## Método numérico para achar o pH

```python
#definindo a função para achar o Vtit para cada pH

def pH(Vtit):
    Ca=0.1 #definindo as constantes
    Cb=0.1
    V0=10
    Ka1=10**(-3.6)
    Ka2=10**(-6.3)
    Kw=1e-14
    def pHH(ph):
        H=10**(-ph)
        OH=Kw/H
        alpha_ha=(H*Ka1)/(H**2+Ka1*H+Ka1*Ka2) #definindo os alphas
        alpha_a2=(Ka1*Ka2)/(H**2+Ka1*H+Ka1*Ka2)
        return Vtit-(V0*(Ca*(alpha_ha+2*alpha_a2)+OH-H))/(H-OH+Cb) #coração
        
    return raizbs(pHH,0,14+log(Cb))
```

# Ácido tiprótico

## Forma numérica para o pH

```python
def pH(vtit):
    Ca=0.047286 #definindo as constantes
    Cb=0.111
    V0=50
    Ka1=10**(-2.16)
    Ka2=10**(-7.21)
    Ka3=10**(-12.32)
    Kw=1e-14
    
    def pHH(ph):
        H=10**(-ph)
        OH=Kw/H
        alpha_h2a= (Ka1*H**2)/(H**3+Ka1*H**2+Ka1*Ka2*H+Ka1*Ka2*Ka3)#definindo os alphas
        alpha_ha=(Ka1*Ka2*H)/(H**3+Ka1*H**2+Ka1*Ka2*H+Ka1*Ka2*Ka3)
        alpha_a= (Ka1*Ka2*Ka3)/(H**3+Ka1*H**2+Ka1*Ka2*H+Ka1*Ka2*Ka3)
        return vtit-(V0*(Ca*(alpha_h2a+2*alpha_ha+3*alpha_a)+OH-H))/(H-OH+Cb) #coração
        
    return raizbs(pHH,0,14+log(Cb))
```

# Titulação de precipitação

$ \displaystyle Ag^2 - \left( \frac{C_{Ag}V_{Ag}- C_{Cl}V_{Cl}}{V_{Ag}+V_{Cl}} \right) Ag - K_{ps}=0$

$\displaystyle Ag \cdot Cl=K_{ps}$

```python
def pAg(vtit):
    CCl=0.05 #definindo as constantes 
    VCl=50
    CAg=0.1
    Kps=1.82e-10
    
    def pagg(pag): #definindo a função 
        Ag=10**(-pag)
        return Ag**2 - ((CAg*vtit-CCl*VCl)/(vtit+VCl))*Ag - Kps
    
    return raizbs(pagg,-log(CAg),14) #achando o pAg  
```

```python
VAg,pag,pcl=[],[],[]
Kps=1.82e-10

for vag in arange(0,50,0.05):
    VAg.append(vag)
    prata=pAg(vag)
    pag.append(prata)
    pcl.append(-log(Kps/10**(-prata)))

plt.plot(VAg,pag,linewidth=1.5,color='black')
plt.plot(VAg,pcl,linewidth=1.5,color='red')
plt.xlabel('$V_{Ag}$',size=15)
plt.title('Curva de titulação do NaCl por AgNO$_3$',size=18)
plt.legend(['pAg','pCl'],fontsize='x-large')
```

# Titulação de complexação

### Antes do PE

$\displaystyle Ca=\frac{C_{Ca}V_{Ca}-C_{EDTA}V_{EDTA}}{V_{Ca}+V_{EDTA}}$

### No PE

$\displaystyle CaY2=  \frac{C_{Ca}V_{Ca}}{V_{Ca}+V_{EDTA}}$

$\displaystyle Ca=\sqrt{\frac{CaY2}{K'_{f}}}$

### Depois do PE

$\displaystyle CaY2=  \frac{C_{Ca}V_{Ca}}{V_{Ca}+V_{EDTA}} \; \; ; \; \; \displaystyle Y4=\frac{C_{EDTA}V_{EDTA}-C_{Ca}V_{Ca}}{V_{EDTA}+V_{Ca}}$

$\displaystyle K'_{f} = \frac{CaY2}{Ca \cdot Y4} \therefore { Ca=\frac{CaY2}{K'_{f}\cdot Y4} }$

```python
def pCa(v_edta):
    CCa=0.005
    VCa=50
    CEDTA=0.01
    Kf=1.75e10
    
    def antes_do_pe(v_edta):
        return -log((CCa*VCa-CEDTA*v_edta)/(VCa+v_edta))
    
    def no_pe(v_edta):
        CaY2=(CCa*VCa)/(VCa+v_edta)
        return -log(raiz(CaY2/Kf))
    
    def depois_do_pe(v_edta):
        CaY2=(CCa*VCa)/(VCa+v_edta)
        Y4=(CEDTA*v_edta - CCa*VCa)/(VCa+v_edta)
        return -log((CaY2)/(Kf*Y4))

    PE=(CCa*VCa)/(CEDTA)
    
    if v_edta<PE:
        return antes_do_pe(v_edta)
    
    elif v_edta==PE:
        return no_pe(v_edta)
    
    else:
        return depois_do_pe(v_edta)
```

```python
pca,vtit=[],[]
for i in arange(0,50,0.05):
    pca.append(pCa(i))
    vtit.append(i)

plt.plot(vtit,pca,linewidth=1,color='black')
plt.xlabel('$V_{EDTA}$',size=15)
plt.ylabel('pCa',size=15)
plt.title('Curva de titulação do Ca$^{2+}$ por EDTA',size=18)
```

# Titulação redox

### Titulação com o redutor na bureta

$ \displaystyle E_{ref}=0,241$ V (calomelano)

$\displaystyle \Delta E= E-E_{ref}$

### Antes do PE

$\displaystyle E=E_0 - 0.0592\cdot \log\left(\frac{n_{\text{Fe}^{2+}}/V_T}{n_{\text{Fe}^{3+}}/V_T}\right)$  

$\displaystyle n_{\text{Fe}^{3+}}=n_{\text{Ce}^{4+}} \therefore n_{\text{Fe}^{3+}} = V_{Ce^{4+}}\cdot C_{Ce^{4+}}$

$\displaystyle n_{\text{Fe}^{2+}}=C_{\text{Fe}^{2+}} V_{\text{Fe}^{2+}} - n_{\text{Fe}^{3+}}$

### No PE

$\displaystyle E=\frac{E_{\text{Fe}}+E_{\text{Ce}}}{2}$

### Depois do PE

$\displaystyle E=E_0 - 0.0592\cdot \log\left(\frac{n_{\text{Ce}^{3+}}/V_T}{n_{\text{Ce}^{4+}}/V_T}\right)$

$\displaystyle n_{\text{Ce}^{3+}} = C_{\text{Fe}^{2+}} V_{\text{Fe}^{2+}}$

$\displaystyle n_{\text{Ce}^{4+}} = C_{\text{Ce}^{4+}}V_{\text{Ce}^{4+}}-n_{\text{Ce}^{3+}}$

### Se quiser fazer com o oxidante na bureta basta fazer as contas de baixo para cima

```python
def ddp(VCe):
    if VCe==0:
        raise ValueError('Vtit não pode ser igual a zero.')
        
    E0_Fe=0.767 #Fe3+ + e- -> Fe2+
    E0_Ce=1.7 #Ce4+ + e- -> Ce3+
    E_ref=0.241 #eletrodo de calomelano
    CCe=0.1
    CFe2=0.1
    VFe2=25
    
    def antes_do_pe(VCe):
        Fe3=VCe*CCe
        Fe2=VFe2*CFe2-Fe3
        return E0_Fe-0.0592*log(Fe2/Fe3) -E_ref
    
    def no_pe(VCe):
        E=(E0_Fe+E0_Ce)/2
        return E-E_ref
        
    def depois_do_pe(VCe):
        Ce3=VFe2*CFe2
        Ce4=CCe*VCe-Ce3
        return E0_Ce-0.0592*log(Ce3/Ce4) -E_ref
    
    PE=(CFe2*VFe2)/(CCe)
    
    if VCe<PE:
        return antes_do_pe(VCe)

    elif VCe==PE:
        return no_pe(VCe)
    
    else:
        return depois_do_pe(VCe)
```

```python
E,vtit=[],[]
for i in arange(5,40,0.15):
    E.append(ddp(i))
    vtit.append(i)

plt.plot(vtit,E,linewidth=1,color='black')
plt.xlabel('V$_{Ce^{4+}}$',size=15)
plt.ylabel('$\Delta$E',size=15)
plt.title('Curva de titulação do Fe$^{2+}$',size=18)
```
